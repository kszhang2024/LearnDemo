## 题目二：提取用户信息更新表和登录信息表的时间信息

#### 用户登录信息表Training_LogInfo.csv中均存在大量的时间数据，提取时间数据内存在的信息，一方面可以加深对数据的理解，另一方面能够探索这部分信息和目标的关联程度。同时用户登录时间（‘LogInfo3’）、借款成交时间（‘Listinginfo1’）这些时间的时间差信息也能够反映出P2P网络贷款不同用户的行为信息。

```python
data1 = pd.read_csv('data/Training_LogInfo.csv')
data1[:5]
```

##### 使用to_datetime函数转换用户登录信息表的时间字符串。

```python
data1['LogInfo3'] = pd.to_datetime(data1['LogInfo3'])
data1['Listinginfo1'] = pd.to_datetime(data1['Listinginfo1'])
data1[:5]
```

##### 新增3列，取值分别为字段‘Listinginfo1’的时间信息：年、月、周。

```python
year = {}
month = {}
week = {}
data1['year'] =data1['Listinginfo1'].dt.year
data1['month'] =data1['Listinginfo1'].dt.month
data1['week'] =data1['Listinginfo1'].dt.isocalendar().week
data1[:5]
```

##### 计算用户登录信息表中两时间的差，分别以日、小时、分钟计算。

```python
day = {}
hour = {}
sces = {}
for i in range (len(data1['Listinginfo1'])):
    day[i] = data1['Listinginfo1'][i]-data1['LogInfo3'][i]
    sces[i] = day[i].total_seconds()
    hour[i] = sces[i]/60/60
```

```python
print(day[0])
print(sces[0])
print(hour[0])
```

## 题目三：使用分组聚合方法进一步分析用户信息更新表和登录信息表

#### 分析用户信息更新表和登录信息表时，除了提取时间本身的信息外，还可以结合用户编号进行分组聚合，然后进行组内分析。通过组内分析可以得出每组组内的最早和最晚信息更新时间、最早和最晚登录时间、信息更新的次数、登录的次数等信息。

##### 使用groupby方法对用户登录信息表按用户Id进行分组。

```python
groups = data1.groupby(['Idx'])
groups
```

```python
groups.get_group(10001)
```

##### 使用agg方法求取分组后的最早和最晚登录时间与最早借款成交时间，并使用apply函数在每一行上计算最早借款成交时间与最早登录时间的差值。

```python
df = groups.agg({'LogInfo3':['min','max'],'Listinginfo1':['min']})
df
```

```python
def get_time(df):
    return df['Listinginfo1']['min'] - df['LogInfo3']['min']
df1 = df.apply(get_time,axis = 1)
df1
```

## 题目四：使对用户信息更新表和登录信息表进行长宽表转换

#### 通过对数据的描述性统计，以及时间数据信息提取，分组聚合操作已经获得了相当多的信息，但用户信息更新表Training_Userupdate.csv和登录信息表Training_LogInfo.csv是长表，而主表是宽表，需要通过长宽表转换将数据合并在一张表内。

##### 使用povit_table函数进行长宽表转换：
登录信息表Training_LogInfo.csv以'Idx'为行索引，列索引为'LogInfo2'，值为‘LogInfo1’，聚合函数为count生成透视表。


```python
dd = pd.pivot_table(data1,index='Idx',columns='LogInfo2',
                    values = 'LogInfo1',aggfunc='count')
dd
```

##### 使用crosstab函数进行长宽表转换：
登录信息表Training_LogInfo.csv以'Idx'为行索引，列索引为'LogInfo2'，值为‘LogInfo1’，聚合函数为count，生成交叉表。


```python
cc = pd.crosstab(index=data1['Idx'],columns=data1['LogInfo2'],
                values=data1['LogInfo1'],aggfunc='count')
cc
```
